```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import json
import requests
import alpaca_trade_api as tradeapi
```

```python
from dotenv import load_dotenv
load_dotenv('.env')
```

```python
alpaca_api_key=os.getenv('ALPACA_API_KEY')
alpaca_secret_key=os.getenv('ALPACA_SECRET_KEY')
```

```python
api=tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version='v2'
)
```

```python
t1=['HMC','UL','BBY','BLNK','ANTM','KO','CTVA','CVS','EPC','ETSY','EXC','F','FOSL','FMS']
t2=['GME','CL','HP','SJM','EPC','VFC','SIG','LEVI','LULU','MSFT','TAP','OC','PEP','PFE','RL','RDFN']
t3= ['SAM','SAP','BNS','SIX','SBUX','TGT','TD','CLX','CZR','HSY','VFC','UPWK','VZ','VRTX','FB','WBA','WSFS']
t4='SPY' # have not run this don't know the ticker. I think we may need this to see how the broadr mkt moved during that period
```

```python
timeframe='1D'
start_date=pd.Timestamp("2020-06-01",tz='America/New_York').isoformat()
end_date=pd.Timestamp("2020-12-31",tz='America/New_York').isoformat()

ticker_data4=api.get_barset(
    t4,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,

).df
```

```python

timeframe='1D'
start_date=pd.Timestamp("2020-06-01",tz='America/New_York').isoformat()
end_date=pd.Timestamp("2020-12-31",tz='America/New_York').isoformat()

ticker_data1=api.get_barset(
    t1,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,

).df
```

```python
timeframe='1D'
start_date=pd.Timestamp("2020-06-01",tz='America/New_York').isoformat()
end_date=pd.Timestamp("2020-12-31",tz='America/New_York').isoformat()

ticker_data2=api.get_barset(
    t2,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,

).df
```

```python
timeframe='1D'
start_date=pd.Timestamp("2020-06-01",tz='America/New_York').isoformat()
end_date=pd.Timestamp("2020-12-31",tz='America/New_York').isoformat()

ticker_data3=api.get_barset(
    t3,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,

).df
```

```python
# Create and empty DataFrame for closing prices
df_closing_prices3 = pd.DataFrame()

# Fetch the closing prices for all the tickers
for ticker in t3:
    df_closing_prices3[ticker] = ticker_data3[ticker]["close"]

# Drop the time component of the date
df_closing_prices3.index = df_closing_prices3.index.date

# Display sample data
df_closing_prices3.head()
```

```python
# Create and empty DataFrame for closing prices
df_closing_prices2 = pd.DataFrame()

# Fetch the closing prices for all the tickers
for ticker in t2:
    df_closing_prices2[ticker] = ticker_data2[ticker]["close"]

# Drop the time component of the date
df_closing_prices2.index = df_closing_prices2.index.date

# Display sample data
df_closing_prices2.head()
```

```python
df_closing_prices1 = pd.DataFrame()

# Fetch the closing prices for all the tickers
for ticker in t1:
    df_closing_prices1[ticker] = ticker_data1[ticker]["close"]

# Drop the time component of the date
df_closing_prices1.index = df_closing_prices1.index.date

# Display sample data
df_closing_prices1.head()
```

```python
df_closing_prices4 = pd.DataFrame()
df_closing_prices4['SPY'] = ticker_data4['SPY']["close"]

df_closing_prices4.index = df_closing_prices4.index.date

# Display sample data
df_closing_prices4.head()
```

```python
combined_ticker_clsprices=pd.concat([df_closing_prices1,df_closing_prices2,df_closing_prices3,df_closing_prices4],axis=1,join='inner')
```

```python
combined_ticker_clsprices
```

```python
change_prices=combined_ticker_clsprices.pct_change().dropna()
change_prices.head()
change_prices.to_csv('change_prices_2020.csv')
```
