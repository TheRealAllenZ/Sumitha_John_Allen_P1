```python

# Can try with small list.

for item in company_list1:
    qtr= requests.get(f'https://fmpcloud.io/api/v3/income-statement/{item}?period=quarter&limit=400&apikey=727d30427f4f7a794f3996053d9ef4c5')
    data=qtr.json()
    print(f'{item} is processing/n') # if line is printed and an error is shown then remove company from the list
    Q1_date=data[0]['date']
    Q1_Revenue = data[0]['revenue']
    Q1_EBITDA=data[0]['ebitda']
    Q1_Netincome= data[0]['netIncome']

    Q2_date=data[1]['date']
    Q2_Revenue = data[1]['revenue']
    Q2_EBITDA=data[1]['ebitda']
    Q2_Netincome= data[1]['netIncome']

    Q3_date=data[2]['date']
    Q3_Revenue = data[2]['revenue']
    Q3_EBITDA=data[2]['ebitda']
    Q3_Netincome= data[0]['netIncome']

    Q4_date=data[3]['date']
    Q4_Revenue = data[3]['revenue']
    Q4_EBITDA=data[3]['ebitda']
    Q4_Netincome= data[3]['netIncome']

    Q5_date=data[4]['date']
    Q5_Revenue = data[4]['revenue']
    Q5_EBITDA=data[4]['ebitda']
    Q5_Netincome= data[4]['netIncome']

    f={
    'date':[Q1_date,Q2_date,Q3_date,Q4_date,Q5_date],
    'Revenue':[Q1_Revenue,Q2_Revenue,Q3_Revenue,Q4_Revenue,Q5_Revenue],
    'EBITDA':[Q1_EBITDA,Q2_EBITDA,Q3_EBITDA,Q4_EBITDA,Q5_EBITDA],
    'Net_income':[Q1_Netincome,Q2_Netincome,Q3_Netincome,Q4_Netincome,Q5_Netincome]
    }
    df1=pd.DataFrame(data=f)
    df1.insert(0,'Company', [f'{item}',f'{item}',f'{item}',f'{item}',f'{item}'], True)
    df1.to_csv(f'output_final/{item}.csv')
```
